### Sums of Functions

---

#### Q. Plot a graph of $f(x) + g(x)$ as a <b><u>solid blue line</u></b>.

```python
plt.plot(x, f(x) + g(x), 'b-')
plt.show()
```

#### Q. Describe how the two sinusoids interact when you add them. Answer in 1 to 2 sentences.

When the two functions are added, the result combines the fast oscillations of 

(

)
f(x) with the slower wave pattern of 

(

)
g(x). The slower function acts like a smooth baseline, while the faster function creates rapid fluctuations around it.

### Products of Functions

---

#### Q. Plot a graph of $f(x)g(x)$ as a <b><u>solid red line</u></b>. (1 pt.)

```python
plt.plot(x, f(x) * g(x), 'r-')
plt.show()
```

#### Q. Describe how the two sinusoids interact when you multiply them. Answer in 1 to 2 sentences.

When multiplied, the slower sinusoid modulates the amplitude of the faster one. This produces a wave that looks like a high-frequency signal whose peaks grow and shrink according to the slower function.

(

(

)
f(x), which distorts the regular sinusoidal pattern. The result is a more irregular oscillation where the frequency changes depending on the value of 

(

)

### `fit_data2.csv`

---

#### Q. Load the `fit_data2.csv` dataset using pandas and print the dataframe.

```python
df_data2 = pd.read_csv('fit_data2.csv')
df_data2
```

#### Q. Create two variables `x` and `y` and assign the corresponding dataframe columns to them. Convert `x` and `y` into numpy arrays.

```python
x = df_data2['x']
y = df_data2['y']   

x = x.to_numpy()
y = y.to_numpy()    
```

#### Q. Plot a graph of the data as a <b><u>dashed blue line</u></b>.

```python
plt.plot(x,y,'b--')
plt.show()
```

#### Q. Based on the plot above, create your own custom function to fit the data. Print the optimal parameters and SSR. 

```python
def custom_model(x, a, b):
    return a * x * np.exp(-b * x)

popt, pcov = curve_fit(custom_model, x, y)
a_fit, b_fit = popt

residuals = y - custom_model(x, a_fit, b_fit)
ssr = np.sum(residuals**2)

print('a =', str(popt[0]))
print('b =', str(popt[1]))
print('SSR =', str(ssr))
```

#### Q. Plot the data as <u><b>blue dots</u></b> and plot you fitted function as a <b><u>smooth solid red line</u></b>. 

```python
plt.plot(x, y, 'b.', label='fit_data2') 
plt.plot(x, custom_model(x, *popt), 'r-', label='Custom Fit') 
plt.legend()
plt.show()
```
